### Фильтрация данных


```python
users = users[(users['bd'] >= 10) & (users['bd'] <= 100)]
```

```python
train_df = train_df[train_df['msno'].isin(users['msno'])]
```

```python
songs = songs[songs['song_length'] >= 2000]
```

```python
train_df = train_df[train_df['song_id'].isin(songs['song_id'])]
```

### Feature engineering

Добавляем в train_df и test_df столбцы city, bd, gender из таблицы users и столбцы song_length, genre_ids, artist_name, composer, lyricist, language из таблицы songs

```python
train_df = pd.merge(train_df, users, on='msno')
train_df = train_df.drop(
    ['source_system_tab', 'source_screen_name', 'source_type', 'registered_via', 'registration_init_time', 'expiration_date'], 
                         axis=1)
train_df = pd.merge(train_df, songs, on='song_id')
```

```python
train_df.gender.fillna(value='undefined', inplace=True)
train_df.language.fillna(value='0.0', inplace=True)
train_df.genre_ids.fillna(value='0.0', inplace=True)
train_df.composer.fillna(value='undefined', inplace=True)
train_df.lyricist.fillna(value='undefined', inplace=True)
```

```python
train_df['language'] = train_df['language'].astype(float)
train_df['language'] = train_df['language'].astype(int)
```

Добавляем новый признак в train_df: средняя длина песен, прослушанных пользователем

```python
average_song_length = train_df.groupby('msno')['song_length'].mean().reset_index()

train_df = pd.merge(train_df, average_song_length, on='msno', suffixes=('', '_avg_song_length'))
```

Теперь для каждой песни найдём её популярность, т.е. число пользователей, прослушавших её

```python
song_listen_counts = train_df['song_id'].value_counts().reset_index()
song_listen_counts.columns = ['song_id', 'listen_count']

train_df = pd.merge(train_df, song_listen_counts, on='song_id', how='left')
```

Удалим дубликаты из датафреймов train_df и test_df

```python
train_df = train_df.drop_duplicates().reset_index(drop=True)
```
